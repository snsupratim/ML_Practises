```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import model_selection
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("dataset.csv")
```

```python
data.shape
```

```python
data.head(1)
```

```python
meddata=data[['Age','Blood Pressure','Cholesterol Levels','Heart Rate','BMI','Diagnosis']]
meddata.head(1)
```

```python
meddata['Diagnosis'] = meddata['Diagnosis'].apply(
    lambda x: 1 if any(condition in x for condition in ['Hypertension', 'Obesity', 'Overweight']) else 0
)
```

```python
meddata.head()
```

```python
X = meddata.drop('Diagnosis', axis=1)
y = meddata['Diagnosis']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20)
```

```python
#process1

num_trees=10
model=AdaBoostClassifier(n_estimators=num_trees)
```

```python
kfold=model_selection.KFold(n_splits=6)
results=model_selection.cross_val_score(model,X_train,y_train,cv=kfold)
```

```python
print(results)
```

```python
print(results.mean())
```

```python
X = meddata.drop('Diagnosis', axis=1)
y = meddata['Diagnosis']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20)
```

```python
#process2
model=AdaBoostClassifier(n_estimators=num_trees)
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```
