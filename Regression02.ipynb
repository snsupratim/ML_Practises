```python
import pandas as pd
import numpy as np
import matplotlib.pyplot  as plt
```

```python
df=pd.read_csv("USA_Housing.csv")
df.head()
```

```python
dataset=df[['Avg. Area Income','Avg. Area House Age','Avg. Area Number of Rooms','Avg. Area Number of Bedrooms',,'Area Population''Price']].copy()
dataset_columns=
dataset.head()
```

```python
X=dataset[['Avg. Area Income','Avg. Area House Age','Avg. Area Number of Rooms','Avg. Area Number of Bedrooms']]
Y=dataset['Price']
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.20)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
print(lr.score(x_test,y_test))
```
