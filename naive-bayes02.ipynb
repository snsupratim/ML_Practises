```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn
```

```python

dataset = pd.read_csv('data-2.csv')
dataset.head()
```

```python
import pandas as pd



# Separate the 'diagnosis' column as the target (y) and the rest as features (X)
# Assuming 'id' is the first column, 'diagnosis' is the second column
X = dataset.iloc[:, 2:-1]  # Select all columns except 'id' and 'diagnosis'
y = dataset.iloc[:, 1]   # Select only the 'diagnosis' column as the target
```

```python
X
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

```python
y_pred  =  classifier.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm = confusion_matrix(y_test, y_pred)
ac = accuracy_score(y_test,y_pred)
```
